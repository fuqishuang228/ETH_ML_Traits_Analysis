```python
import pandas as pd
import numpy as np
import os
import csv
```

```python
#Event='AscendEx'
Event='Upbit'
raw_path='../Dataset/'+Event+'/all_data_token/'
address_path='../Dataset/'+Event+'/dataset/'
account_feature_output_path = '../Code_Results/Intermediary_Account_Feature_Results/'+Event+'_account_feature.csv'
feature_output_path ='../Code_Results/Intermediary_Account_Feature_Results/'+Event
```

```python
#Step1：筛选出之发生了ETH交易的洗钱账户（source+layering）
#方法2：all_erc20_address和all_normal_address的交集，再用all_normal_address减掉交集
all_address=pd.read_csv(address_path+"all-address.csv")
all_erc20_address=pd.read_csv(address_path+"all-erc20-address.csv")
all_normal_address=pd.read_csv(address_path+"all-normal-address.csv")
address=all_address['address']
erc20_address=all_erc20_address['address']
normal_address=all_normal_address['address']
ETH_address=set(address)-set(erc20_address)
temp=set(erc20_address)&set(normal_address)
ETH_address=list(set(normal_address)-temp)
#再筛选出来label是heist的账户
all_ETH_address=all_normal_address[all_normal_address['address'].isin(ETH_address)]
ETH_heist_address=list(all_ETH_address.loc[all_ETH_address['label']=='heist']['address'])
```

```python
len(ETH_heist_address)
```

```python
#Step2：筛选出ETH账户的交易
# ETH_address={'0x70dcf33ca09bd87bb2a301280331406ebd32c8a0'}  测试数据
all_normal_tx=pd.read_csv(address_path+"all-normal-tx.csv")
ETH_tx=all_normal_tx[all_normal_tx['from'].isin(list(ETH_address)) | all_normal_tx['to'].isin(list(ETH_address))]
```

```python
with open(account_feature_output_path,'w') as f:

    csv_write = csv.writer(f)
    csv_head = ["address","In","Out","Balance"]
    csv_write.writerow(csv_head)

output = pd.read_csv(account_feature_output_path)
output['address'] = ETH_heist_address
```

```python
#Step3：统计每个账户的余额
for i in range(0,len(ETH_address)):
    record=ETH_tx.loc[ETH_tx["isError"]!=1]

    # 计算转出交易
    #from_cnt = record[(record['from'] == List[i]) & (record['value'] != '0')]
    from_cnt = record[(record['from'] == ETH_address[i])]
    from_ETH_cnt=from_cnt.loc[pd.isnull(from_cnt["contractAddress"])]
    out_value = from_ETH_cnt['value'].astype(float).sum() / (1E+18)

#     # 计算账户转出交易时使用的Gas费
#     Gas_Tx=record_raw.loc[(record_raw['from'] == List[i])& pd.notnull(record_raw['isError'])]

#     #因为内部交易和外部交易会公用一个hash，内部交易的isError是空，所以把内部交易删了
#     from_gas=Gas_Tx.loc[pd.notnull(Gas_Tx['gasPrice']),['gasPrice','gasUsed']]
#     gas_ETH=sum(from_gas['gasPrice']*from_gas['gasUsed'])/ (1E+18)

    # 计算接收交易
    to_cnt = record[(record['to'] == ETH_address[i]) & (record['value']!='0')]  # to:receive
    to_ETH_cnt = to_cnt.loc[pd.isnull(to_cnt["contractAddress"])]
    in_value = to_cnt['value'].astype(float).sum() / (1E+18)
    Balance= in_value-out_value
#     Balance= in_value-out_value-gas_ETH
    output['In'][i] = in_value
    output['Out'][i] = out_value
#     output['Gas_ETH'][i] = gas_ETH
    output['Balance'][i] = Balance
#     output['In_ETH'] = in_value
#     output['Out_ETH'] = out_value
# #     output['Gas_ETH'] = gas_ETH
#     output['Balance']= Balance
```

```python
account_info1 = pd.read_csv('../Dataset/' + Event + '/reference_list/accounts-hacker.csv')
intermediary = account_info1.loc[
    (account_info1['label'] == 'heist') & (account_info1['name_tag'] != 'ml_transit_0')]
intermediary_list = list(intermediary['address'])

# output['address'] = intermediary_list
cnt = len(intermediary_list)

List=intermediary_list
```

```python
cnt
```

```python
for i in range(cnt):
    #print(str(i)+" "+List[i])
    file_address = List[i]
    if file_address not in List:  # 检查地址是否在文件夹中，仅用于有账户列表情况下检查
        print(file_address + " is not in " + raw_path)
    else:
        file_name = raw_path + file_address + '.csv'

        # record = pd.read_csv(file_name,engine= 'python')
        if os.path.exists(file_name):
            record_raw = pd.read_csv(file_name)
        # 读取账户交易文件

        with open(feature_output_path+file_address+'_account_feature.csv','w') as f:

            csv_write = csv.writer(f)
            csv_head = ["address","In_ETH","Out_ETH",'Gas_ETH',"Balance_ETH"]
            csv_write.writerow(csv_head)

        output = pd.read_csv(feature_output_path+file_address+'_account_feature.csv')
        output['address']=file_address
        
        #-------------------------------筛选出ETH————————————————————————
        #--------------------除去交易失败的交易------------------#
        record=record_raw.loc[record_raw["isError"]!=1]

        # 计算转出交易
        #from_cnt = record[(record['from'] == List[i]) & (record['value'] != '0')]
        from_cnt = record[(record['from'] == List[i])]
        from_ETH_cnt=from_cnt.loc[pd.isnull(from_cnt["contractAddress"])]
        out_value = from_ETH_cnt['value'].astype(float).sum() / (1E+18)
        
        # 计算账户转出交易时使用的Gas费
        Gas_Tx=record_raw.loc[(record_raw['from'] == List[i])& pd.notnull(record_raw['isError'])]
        
        #因为内部交易和外部交易会公用一个hash，内部交易的isError是空，所以把内部交易删了
        from_gas=Gas_Tx.loc[pd.notnull(Gas_Tx['gasPrice']),['gasPrice','gasUsed']]
        gas_ETH=sum(from_gas['gasPrice']*from_gas['gasUsed'])/ (1E+18)
        
        # 计算接收交易
        to_cnt = record[(record['to'] == List[i]) & (record['value']!='0')]  # to:receive
        to_ETH_cnt = to_cnt.loc[pd.isnull(to_cnt["contractAddress"])]
        in_value = to_cnt['value'].astype(float).sum() / (1E+18)
        Balance= in_value-out_value-gas_ETH
#         output['In_ETH'][i] = in_value
#         output['Out_ETH'][i] = out_value
#         output['Gas_ETH'][i] = gas_ETH
#         output['Balance'][i] = Balance
        output['In_ETH'] = in_value
        output['Out_ETH'] = out_value
        output['Gas_ETH'] = gas_ETH
        output['Balance']= Balance
        
#         if Balance<0:
#             output['Balance_ETH'][i] = 0
#         else:
#             output['Balance_ETH'][i] = Balance
        #---------------------------------------------------------------
        #----------------筛选出该账户的内部交易---------------#
        Internal_Tx=record.loc[pd.isnull(record['isError'])]
        #得到相关代币的contractAddr（不重复）
        Token_Addr=list(set(Internal_Tx['contractAddress']))
        for token in Token_Addr:
            #筛选和该代币相关的内部交易
            token_Tx=Internal_Tx.loc[Internal_Tx['contractAddress']==token]
            #计算转出交易的token总和
            from_token=token_Tx.loc[token_Tx['from'] == List[i],'value'].astype(float).sum()
            #计算接受交易的token总和
            to_token=token_Tx.loc[token_Tx['to'] == List[i],'value'].astype(float).sum()
            #计算from和To差值
            balance_token=to_token-from_token
            token_symnbol=list(Internal_Tx.loc[Internal_Tx['contractAddress']==token,'tokenSymbol'])[0]
            #判断，这个代币相关的余额是否为0：
            if balance_token !=0:
                 output['Balance_'+token_symnbol] = balance_token
#                 if token_symnbol in output.columns:
#                     output['Balance_'+token_symnbol][i] = balance_token
                   
#                 else:
#                     output['Balance_'+token_symnbol]= 0#先创建
#                     output['Balance_'+token_symnbol][i] = balance_token#再赋值

        break
```
